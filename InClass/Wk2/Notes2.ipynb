In Class Notes: Pivoting

```python
#import packages
import pandas as pd
#importing the data
population = pd.read_csv("/Users/stolieerickson/Documents/MSBA/Fall/GSB544-Computing/InClass/Wk2/pop.csv")
#pivoting
long_population = population.melt(id_vars=["country"], var_name="year", value_name="population")
#import other dataset
co2 = pd.read_csv("/Users/stolieerickson/Documents/MSBA/Fall/GSB544-Computing/InClass/Wk2/co2.csv")
#pivoting
long_co2 = co2.melt(id_vars=["country"], var_name="year", value_name="emissions")
#joining
long_population.merge(long_co2, on=["country", "year"])

```
